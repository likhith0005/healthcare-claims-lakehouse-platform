```python
from pyspark.sql.functions import *
from pyspark.sql.types import *
import time

print("=" * 60)
print("REAL TIME STREAMING FRAMEWORK")
print("=" * 60)
stream_batch_1 = [

    (1001, "HEART_ATTACK", "CRITICAL"),
    (1002, "DIABETES", "NORMAL"),
    (1003, "PNEUMONIA", "HIGH")

]

schema = StructType([

    StructField("patient_id", IntegerType(), True),
    StructField("disease", StringType(), True),
    StructField("severity", StringType(), True)

])

stream_df_1 = spark.createDataFrame(stream_batch_1, schema)
print("\n[STREAM] Batch 1 Received")
display(stream_df_1)
time.sleep(2)
analytics_1 = stream_df_1.groupBy("severity").count()
display(analytics_1)
print("[SUCCESS] Stream Batch 1 Processed")
stream_batch_2 = [

    (1004, "STROKE", "CRITICAL"),
    (1005, "HYPERTENSION", "MEDIUM"),
    (1006, "COVID", "CRITICAL")

]

stream_df_2 = spark.createDataFrame(stream_batch_2, schema)
print("\n[STREAM] Batch 2 Received")
display(stream_df_2)
time.sleep(2)
analytics_2 = stream_df_2.groupBy("severity").count()
display(analytics_2)
print("[SUCCESS] Stream Batch 2 Processed")
final_stream_df = stream_df_1.union(stream_df_2)
display(final_stream_df)
critical_alerts = final_stream_df.filter(
    col("severity") == "CRITICAL"
)

display(critical_alerts)

print("[ALERT] Critical Patients Identified")ream_df.write \
    .format("delta") \
    .mode("overwrite") \
    .saveAsTable("healthcare_migration.gold.real_time_streaming_events")
print("[SUCCESS] Streaming Results Saved")
print("=" * 60)
print("REAL TIME STREAMING COMPLETED")
print("=" * 60)
```
