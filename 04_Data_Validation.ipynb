```python
from pyspark.sql.functions import col, count

bronze_df = spark.read.table("healthcare_migration.bronze.claims_cleaned")
silver_df = spark.read.table("healthcare_migration.silver.claims_silver")

bronze_count = bronze_df.count()
silver_count = silver_df.count()

print(f"Bronze Count : {bronze_count}")
print(f"Silver Count : {silver_count}")

null_patientid = silver_df.filter(
    col("PATIENTID").isNull()
).count()

print(f"Null PATIENTID Count : {null_patientid}")

duplicate_count = (
    silver_df.groupBy("Id")
    .agg(count("*").alias("cnt"))
    .filter(col("cnt") > 1)
    .count()
)

print(f"Duplicate Id Count : {duplicate_count}")

if null_patientid == 0 and duplicate_count == 0:
    print("Data Validation PASSED")
else:
    print("Data Validation FAILED")
```
