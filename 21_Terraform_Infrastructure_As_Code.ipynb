```python
print("=" * 60)
print("TERRAFORM INFRASTRUCTURE AS CODE")
print("=" * 60)
terraform_script = """

# ============================================
# PROVIDER CONFIGURATION
# ============================================

provider "databricks" {

  host  = "https://demo.cloud.databricks.com"
  token = "sample-token"

}

# ============================================
# RESOURCE GROUP
# ============================================

resource "databricks_cluster" "healthcare_cluster" {

  cluster_name            = "healthcare-etl-cluster"
  spark_version           = "14.3.x-scala2.12"
  node_type_id            = "Standard_DS3_v2"
  autotermination_minutes = 30

  autoscale {

    min_workers = 1
    max_workers = 4

  }
}

# ============================================
# NOTEBOOK DEPLOYMENT
# ============================================

resource "databricks_notebook" "etl_pipeline" {

  path     = "/Workspace/Healthcare/ETL_Pipeline"
  language = "PYTHON"

}

# ============================================
# JOB ORCHESTRATION
# ============================================

resource "databricks_job" "healthcare_job" {

  name = "healthcare-master-job"

}

"""

print(terraform_script)
components = [

    ("Databricks Cluster", "Provisioned"),
    ("ETL Notebook", "Deployed"),
    ("Job Orchestration", "Configured"),
    ("Auto Scaling", "Enabled"),
    ("Cloud Infrastructure", "Ready")

]

infra_df = spark.createDataFrame(
    components,
    ["component", "status"]
)

display(infra_df)
infra_df.write \
    .format("delta") \
    .mode("overwrite") \
    .saveAsTable("healthcare_migration.gold.terraform_infra_status")

print("[SUCCESS] Infrastructure Status Saved")

print("=" * 60)
print("TERRAFORM IAC COMPLETED")
print("=" * 60)
```
