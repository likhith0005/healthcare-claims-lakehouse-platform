```python
from pyspark.sql.functions import count

gold_df = (
    spark.read.table("healthcare_migration.silver.claims_silver")
    .groupBy("DEPARTMENTID")
    .agg(
        count("*").alias("total_claims")
    )
)

gold_df.write \
    .format("delta") \
    .mode("overwrite") \
    .saveAsTable("healthcare_migration.gold.department_claim_summary")

print("Gold aggregation created successfully")
```
