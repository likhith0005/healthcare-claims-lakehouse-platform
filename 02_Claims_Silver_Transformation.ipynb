```python
from pyspark.sql.functions import col

silver_df = (
    spark.read.table("healthcare_migration.bronze.claims_cleaned")
    .filter(col("PATIENTID").isNotNull())
)

silver_df.write \
    .format("delta") \
    .mode("overwrite") \
    .saveAsTable("healthcare_migration.silver.claims_silver")

print("Silver table created successfully")
```
