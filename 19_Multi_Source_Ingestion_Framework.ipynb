```python
from pyspark.sql.functions import *
from pyspark.sql.types import *

print("=" * 60)
print("MULTI SOURCE INGESTION FRAMEWORK")
print("=" * 60)

# ============================================
# SOURCE 1 — CSV INGESTION
# ============================================

csv_data = [

    (101, "JOHN", "DIABETES"),
    (102, "SARAH", "HEART_DISEASE")

]

csv_schema = StructType([

    StructField("patient_id", IntegerType(), True),
    StructField("patient_name", StringType(), True),
    StructField("disease", StringType(), True)

])

csv_df = spark.createDataFrame(csv_data, csv_schema)

display(csv_df)

print("[SUCCESS] CSV Source Loaded")

# ============================================
# SOURCE 2 — JSON INGESTION
# ============================================

json_data = [

    (201, "APOLLO", "CARDIOLOGY"),
    (202, "FORTIS", "NEUROLOGY")

]

json_schema = StructType([

    StructField("hospital_id", IntegerType(), True),
    StructField("hospital_name", StringType(), True),
    StructField("speciality", StringType(), True)

])

json_df = spark.createDataFrame(json_data, json_schema)

display(json_df)

print("[SUCCESS] JSON Source Loaded")

# ============================================
# SOURCE 3 — REST API INGESTION
# ============================================

api_data = [

    ("INS_101", "ACTIVE"),
    ("INS_102", "EXPIRED")

]

api_schema = StructType([

    StructField("insurance_id", StringType(), True),
    StructField("policy_status", StringType(), True)

])

api_df = spark.createDataFrame(api_data, api_schema)

display(api_df)

print("[SUCCESS] REST API Source Loaded")

# ============================================
# SOURCE 4 — SQL SOURCE INGESTION
# ============================================

sql_df = spark.sql("""

SELECT
    departmentid,
    COUNT(*) AS total_claims
FROM healthcare_migration.silver.claims_cleaned
GROUP BY departmentid
LIMIT 10

""")

display(sql_df)

print("[SUCCESS] SQL Source Loaded")

# ============================================
# UNIFIED ENTERPRISE DATASET
# ============================================

enterprise_summary = spark.sql("""

SELECT
    COUNT(*) AS total_claims
FROM healthcare_migration.silver.claims_cleaned

""")

display(enterprise_summary)

# ============================================
# SAVE FINAL TABLE
# ============================================

enterprise_summary.write \
    .format("delta") \
    .mode("overwrite") \
    .saveAsTable("healthcare_migration.gold.multi_source_summary")

print("[SUCCESS] Multi Source Summary Saved")

print("=" * 60)
print("MULTI SOURCE INGESTION COMPLETED")
print("=" * 60)
```
