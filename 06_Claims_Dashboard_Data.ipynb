```python
from pyspark.sql.functions import sum, avg, count

dashboard_df = (
    spark.read.table("healthcare_migration.silver.claims_silver")
    .groupBy("DEPARTMENTID")
    .agg(
        count("*").alias("total_claims"),
        avg("PAYER_COVERAGE").alias("avg_coverage"),
        sum("PAYER_COVERAGE").alias("total_coverage")
    )
)

dashboard_df.write \
    .format("delta") \
    .mode("overwrite") \
    .saveAsTable("healthcare_migration.gold.claims_dashboard")

print("Dashboard dataset created successfully")

display(
    spark.read.table(
        "healthcare_migration.gold.claims_dashboard"
    )
)
```
