```python
from pyspark.sql.functions import *
from delta.tables import *
source_df = spark.table(
    "healthcare_migration.silver.claims_cleaned"
)
display(source_df.limit(5))
source_df.write \
    .format("delta") \
    .mode("overwrite") \
    .saveAsTable(
        "healthcare_migration.gold.claims_incremental_target"
    )
print("Initial target table created")
```

```python
cdc_df = source_df.limit(5) \
    .withColumn(
        "PRIMARYPATIENTINSURANCEID",
        lit("UPDATED_INSURANCE")
    )
display(cdc_df)
```

```python
cdc_df.createOrReplaceTempView("claims_cdc_updates")
print("CDC temp view created")
```

```python
merge_sql = """

MERGE INTO healthcare_migration.gold.claims_incremental_target AS target

USING claims_cdc_updates AS source

ON target.Id = source.Id

WHEN MATCHED THEN
UPDATE SET
target.PRIMARYPATIENTINSURANCEID =
source.PRIMARYPATIENTINSURANCEID

WHEN NOT MATCHED THEN
INSERT *

"""

spark.sql(merge_sql)

print("MERGE INTO completed successfully")
```

```python
result_df = spark.sql("""

SELECT
    Id,
    PRIMARYPATIENTINSURANCEID
FROM healthcare_migration.gold.claims_incremental_target
WHERE PRIMARYPATIENTINSURANCEID = 'UPDATED_INSURANCE'

""")

display(result_df)

print("Incremental merge validation completed")
```

```python
incremental_audit_df = spark.sql(f"""

SELECT
    'claims_cleaned' AS source_table,
    'claims_incremental_target' AS target_table,
    {cdc_df.count()} AS updated_records,
    current_timestamp() AS merge_timestamp

""")

display(incremental_audit_df)
```
