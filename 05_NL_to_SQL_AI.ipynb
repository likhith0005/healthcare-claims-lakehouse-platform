```python
user_question = "Show total claims by department"

if "total claims by department" in user_question.lower():

    sql_query = """
    SELECT
        DEPARTMENTID,
        total_claims
    FROM healthcare_migration.gold.department_claim_summary
    ORDER BY total_claims DESC
    """

elif "top department" in user_question.lower():

    sql_query = """
    SELECT
        DEPARTMENTID,
        total_claims
    FROM healthcare_migration.gold.department_claim_summary
    ORDER BY total_claims DESC
    LIMIT 1
    """

else:
    sql_query = "SELECT 'No matching query found'"

print("Generated SQL Query:")
print(sql_query)

result_df = spark.sql(sql_query)

display(result_df)
```
