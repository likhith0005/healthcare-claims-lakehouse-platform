```python
from pyspark.sql.functions import *
from pyspark.sql.types import *

claims_df = spark.table(
    "healthcare_migration.silver.claims_cleaned"
)

print("Claims table loaded successfully")

display(claims_df.limit(10))
```

```python
fraud_df = (
    claims_df
    .withColumn(
        "fraud_score",
        when(col("PRIMARYPATIENTINSURANCEID").isNull(), 80)
        .otherwise(20)
    )
    .withColumn(
        "fraud_flag",
        when(col("fraud_score") >= 70, "HIGH_RISK")
        .otherwise("LOW_RISK")
    )
)

print("Fraud features created")

display(
    fraud_df.select(
        "Id",
        "PATIENTID",
        "fraud_score",
        "fraud_flag"
    ).limit(20)
)
```

```python
fraud_summary_df = (
    fraud_df
    .groupBy("fraud_flag")
    .count()
)

display(fraud_summary_df)
```

```python
(
    fraud_df.write
    .format("delta")
    .mode("overwrite")
    .saveAsTable(
        "healthcare_migration.gold.claims_fraud_predictions"
    )
)

print("Fraud prediction table saved successfully")
```

```python
from pyspark.sql.functions import *

ml_audit_df = spark.sql(f"""
SELECT
    'claims_cleaned' AS source_table,
    'claims_fraud_predictions' AS target_table,
    {fraud_df.count()} AS record_count,
    current_timestamp() AS load_timestamp
""")

display(ml_audit_df)
```
